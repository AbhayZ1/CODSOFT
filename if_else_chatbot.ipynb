```python
import re
from datetime import datetime

def chatbot():
    print("Hello! I'm your chatbot. How can I help you today?")

    while True:
        user_input = input("You: ").strip().lower()

        # Exit condition
        if re.search(r"\b(exit|quit|bye|goodbye)\b", user_input):
            print("Chatbot: Goodbye! Have a great day!")
            break

        elif re.search(r"\b(hello|hi|hey)\b", user_input):
            print("Chatbot: Hello! How can I assist you today?")
        
        elif re.search(r"\b(how are you|how's it going)\b", user_input):
            print("Chatbot: I'm just a program, but I'm functioning perfectly. Thanks for asking!")

        elif re.search(r"\b(your name|who are you|what is your name)\b", user_input):
            print("Chatbot: I'm a rule-based chatbot. You can call me Chatbot!")

        elif re.search(r"\bweather\b", user_input):
            print("Chatbot: I can't check the weather right now, but it’s always a good idea to check a weather app!")

        elif re.search(r"\b(help)\b", user_input):
            print("Chatbot: Sure! Let me know what you need help with.")

        elif re.search(r"\btime\b", user_input):
            current_time = datetime.now().strftime("%H:%M:%S")
            print(f"Chatbot: The current time is {current_time}.")

        elif re.search(r"\bdate\b", user_input):
            current_date = datetime.now().strftime("%Y-%m-%d")
            print(f"Chatbot: Today's date is {current_date}.")

        elif re.search(r"\b(fact|trivia)\b", user_input):
            print("Chatbot: Did you know? Honey never spoils. Archaeologists have found pots of honey in ancient tombs!")

        elif re.search(r"\b(joke)\b", user_input):
            print("Chatbot: Why don’t skeletons fight each other? They don’t have the guts!")

        elif re.search(r"\b(what is|calculate)\b.*\d+", user_input):
            numbers = re.findall(r'\d+', user_input)
            if len(numbers) == 2:
                num1, num2 = map(int, numbers)
                print(f"Chatbot: The sum of {num1} and {num2} is {num1 + num2}.")
            else:
                print("Chatbot: I can only calculate sums of two numbers. Please try again.")

        elif re.search(r"\b(quote|advice)\b", user_input):
            print("Chatbot: 'The best time to plant a tree was 25 years ago. The second best time is now.'")

        elif re.search(r"\b(news)\b", user_input):
            print("Chatbot: I can't fetch live news, but I recommend checking your preferred news website or app.")

        elif re.search(r"\b(health|exercise)\b", user_input):
            print("Chatbot: Regular exercise is great for your health! Try to aim for at least 30 minutes of physical activity each day.")

        elif re.search(r"\b(recommend|suggest)\b", user_input):
            print("Chatbot: I'd recommend reading a good book or taking a walk outside to refresh your mind!")

        elif re.search(r"\b(year)\b", user_input):
            current_year = datetime.now().strftime("%Y")
            print(f"Chatbot: The current year is {current_year}.")

        elif re.search(r"\b(technology|tech)\b", user_input):
            print("Chatbot: Did you know? The first computer virus was created in 1983 and was called 'Creeper'.")

        elif re.search(r"\b(robot|AI)\b", user_input):
            print("Chatbot: I’m an AI, which stands for Artificial Intelligence. My job is to assist and answer your questions!")

        elif re.search(r"\b(book|movie)\b", user_input):
            print("Chatbot: If you're looking for a good book, I recommend '1984' by George Orwell. For a movie, 'The Matrix' is a classic!")

        elif re.search(r"\b(hobbies|interests)\b", user_input):
            print("Chatbot: I don't have hobbies, but I can help suggest one! How about photography, painting, or even learning a new language?")

        elif re.search(r"\b(programming|coding|code)\b", user_input):
            print("Chatbot: I can help with coding! If you're learning Python, I suggest checking out the official Python documentation.")

        else:
            print("Chatbot: I'm sorry, I didn't understand that. Could you try rephrasing or asking something else?")

if __name__ == "__main__":
    chatbot()
```
